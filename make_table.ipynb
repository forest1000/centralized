```python
import matplotlib.pyplot as plt
import pandas as pd

# データフレームの作成
data = {
    'Methods': [
        'FixMatch', 'UniMatch','CauSSL', 'BCP', 'ARCO', 
        'AC-MT', 'ABD(Cross Teaching)','MiDSS one','MONA'
    ],
    '# L': ['5%', '5%', '5%', '5%', '5%', 
            '5%', '5%', '5%','5%'],
    'Vendor A': [
        'LV/MYO/RV/AVG/JC/HD95/ASD', '', '', '', '',
          '', '','','',
    ],
    'Vendor B': [
        'LV/MYO/RV/AVG/JC/HD95/ASD', '', '', '', '', 
        '', '','','',
    ],
    'Vendor C': [
        'LV/MYO/RV/AVG/JC/HD95/ASD', '', '', '', '', 
        '', '','','',
    ],
    'Vendor D': [
        'LV/MYO/RV/AVG/JC/HD95/ASD', '', '', '', '',
        '', '','','',
        ],
    'Average' :[
        'LV/MYO/RV/AVG/JC/HD95/ASD', '', '', '', '',
        '', '','','',
    ]
}

df = pd.DataFrame(data)
fig, ax = plt.subplots(figsize=(12, 4))

# 行ごとに色を交互に設定するためのリストを再定義
row_colors = ['#cce6ff' if i % 2 == 0 else '#e6f2ff' for i in range(len(df))]

# テーブルの作成（インデックスを非表示）
table = ax.table(
    cellText=df.values,
    colLabels=df.columns,
    cellLoc='center',
    loc='center',
    colColours=['#add8e6'] * len(df.columns),
    cellColours=[[row_colors[i]] * len(df.columns) for i in range(len(df))]
)
# "Methods"列のセルの幅を広く設定
for (i, j), cell in table.get_celld().items():
    if j == 0:  # "Methods"列の幅を調整
        cell.set_width(0.15)
    elif j == 1:
        cell.set_width(0.1)
    else:
        cell.set_width(0.2)

# フォントサイズやスタイルの設定
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.6, 2.15)
# 枠線を非表示に
ax.axis('off')

# ファイルに保存
plt.savefig('./mnt/data/idea1_semi Method_M&Ms_table_5%.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
import torch
import os

# チェックポイントファイルのパス
checkpoint_path = '/media/morikawa/DataHDD/home/data/segmentation/fundus_dofe/semi/domain_generalization/checkpoints/experiment_client_1/checkpoints/semi_1729308989.817541/best_loss'

# ファイルが存在するか確認
if not os.path.isfile(checkpoint_path):
    print(f"指定されたファイルが存在しません: {checkpoint_path}")
else:
    # チェックポイントをロード
    checkpoint = torch.load(checkpoint_path, map_location='cpu')  # CPU上でロード

    # チェックポイントのキーを確認
    print("チェックポイントに含まれるキー:")
    for key in checkpoint.keys():
        print(f" - {key}")

    # 各項目の詳細を表示（必要に応じて）
    # 例として 'best_test_loss' を表示
    if 'best_test_loss' in checkpoint:
        print(f"\nBest Test Loss: {checkpoint['best_test_loss']}")
    else:
        print("\n'best_test_loss' キーがチェックポイントに存在しません。")

    # 他の情報も確認したい場合は以下のようにアクセス
    # 例えば、エポック数
    if 'epoch' in checkpoint:
        print(f"Epoch: {checkpoint['epoch']}")

    # モデルの状態辞書やオプティマイザの状態辞書は大きいため、サイズを確認するだけに留める
    if 'model_state_dict' in checkpoint:
        print(f"Model State Dict Keys: {list(checkpoint['model_state_dict'].keys())}")
    if 'optimizer_state_dict' in checkpoint:
        print(f"Optimizer State Dict Keys: {list(checkpoint['optimizer_state_dict'].keys())}")
```
